

### A perspective transform code for transforming traffic signes

```python
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib qt

img = mpimg.imread('traffic_two.jpg')

plt.imshow(img)

def transform(img):
    img_size = (img.shape[1], img.shape[0])
    src=np.float32([[26,105],[385,232],[224,527],[276,530]])
    dst=np.float32([[26,105], [397,105],[224,500],[276,500]])
    M = cv2.getPerspectiveTransform(src, dst)
    Minv = cv2.getPerspectiveTransform(dst, src)
    wraped =  cv2.warpPerspective(img,M,img_size, flags=cv2.INTER_LINEAR)
    
    return  wraped
```

```python
%matplotlib inline
warped= transform(img)

f,(ax1, ax2)= plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('Source image')
ax1.imshow(img)
ax2.set_title('Warpped image')
ax2.imshow(warped)
```
